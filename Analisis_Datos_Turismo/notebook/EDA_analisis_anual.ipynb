## Preparación y Carga de Datos

### Librerías

```python
# Tratamiento de datos
import numpy as np
import pandas as pd

# Gráficos
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

# Otros
import json
```

## Carga de Datos

```python
df_año =  pd.read_csv("año.csv")
```

## Análisis Exploratorio de Datos (EDA)

### Análisis Univariante

Histograma

```python
df_año['TURISTAS'].plot(kind='hist', bins=20, alpha=0.7, title='Histograma de Turistas Anual', cmap="viridis")
plt.savefig('hist_turista_anual.png')
plt.show()

df_año['PERNOCTACIONES'].plot(kind='hist', bins=20, alpha=0.7, title='Histograma de Pernoctaciones Anual', cmap="viridis")
plt.savefig('hist_pernoctacion_anual.png')
plt.show()

df_año['ESTANCIA_MEDIA'].plot(kind='hist', bins=20, alpha=0.7, title='Histograma de Estancia Media Anual', )
plt.savefig('hist_estancia_media_anual.png')
plt.show()

plt.savefig('histograma1.png')
```

Scatter Plots

```python
sns.set_palette("viridis")
scatter=sns.scatterplot(data=df_año, x='AÑO', y='TURISTAS')
scatter.set_title ("Scatterplot de Turistas Anuales")
plt.savefig('scatter_turista_anual.png')
plt.show()
```

```python
sns.set_palette("viridis")
scatter=sns.scatterplot(data=df_año, x='AÑO', y='PERNOCTACIONES')
scatter.set_title ("Scatterplot de Pernoctaciones Anuales")
plt.savefig('scatter_pernoctaciones_anual.png')
plt.show()
```

```python
sns.set_palette("viridis")
scatter=sns.scatterplot(data=df_año, x='AÑO', y='ESTANCIA_MEDIA')
scatter.set_title ("Scatterplot de Estancia Media Anual")
plt.savefig('scatter_estancia_media_anual.png')
plt.show()
```

```python
sns.set_palette("viridis")
scatter=sns.scatterplot(data=df_año, x='ESTANCIA_MEDIA', y='TURISTAS')
scatter.set_title ("Scatterplot de Turistas por Estancia Media Anual")
plt.savefig('scatter_estancia_media_turistas_anual.png')
plt.show()
```

### Análisis Multivariante

```python
df_año = df_año.select_dtypes(exclude=['object'])
scatter=sns.heatmap(df_año.corr(), annot=True, annot_kws={"size": 11}, cmap="viridis")
scatter.set_title ("Heatmap Anual")
plt.savefig('heatmap_anual.png')
plt.show()
```
